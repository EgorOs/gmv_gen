```python
p = 7
S = 4
N = p ** S - 1

m = 2
n = S / m

J = p ** m - 1
L = N / (p ** m - 1)



# x^4+x^2+3x+5
h1 = [1, 0, 1, 3, 5]
A = 4
B = [2, 4, 6]
seed = [4, 0, 5, 5]

def generate_max_seq(p, N, S, B, seed):
    max_seq = seed.copy()
    for i in range(N - S):
        new_val = 0
        for j in range(len(B)):
            new_val += max_seq[i+j]*B[j]
        max_seq.append((new_val)%p)
    return max_seq

# list(enumerate(generate_max_seq(p, N, S, B, seed)))[1800:]
        
```

```python
class MSequnce:
    def __init__(self, p, m, S, B, seed, A=0):
        self.p = p
        self.S = S
        self.B = B
        self.seed = seed
        self.N = p ** S - 1
        self.m = m
        self.n = S / m
        self.J = p ** m - 1
        self.L = self.N // (p ** m - 1)
        self.A = A
        self.vector = self._create_M_seq(p, S, B, seed)
        if self.A != 0:
            self.vector = self.shift(self.A)
    
    def _create_M_seq(self, p, S, B, seed):
        max_seq = seed.copy()
        for i in range(self.N - S):
            new_val = 0
            for j in range(len(B)):
                new_val += max_seq[i+j]*B[j]
            max_seq.append((new_val) % p)
        return np.reshape(max_seq, (-1, 1)).astype(int)
    
    def __copy__(self):
        return MSequnce(self.p, self.m, self.S, self.B, self.seed)
    
    def to_matrix(self):
        return np.reshape(self.vector, (self.J,self.L))
    
    def shift(self, t):
        new_seq = self.__copy__()
        new_seq.vector = np.roll(self.vector, t)
        return new_seq
    
    def show_vector(self, transpose=False):
        df = pd.DataFrame(self.vector.T, index=None) if transpose else pd.DataFrame(self.vector, index=None)
        return HTML(df.to_html())
    
    def show_table(self):
        df = pd.DataFrame(self.to_matrix(), index=None)
        return HTML(df.to_html())
    
    def save_as_vector(self, filepath):
        df = pd.DataFrame(self.vector, index=None)
        df.to_excel(filepath, index=False, header=False)
             
    def save_as_table(self, filepath):
        df = pd.DataFrame(self.to_matrix(), index=None)
        df.to_excel(filepath, index=False, header=False)
        
    def save_object(self, filepath):
        with open(filepath, 'wb') as f:
            pickle.dump(self, f)
    
    def save_all(self, name, dirpath):
        fullpath = Path(dirpath)/name
        os.makedirs(fullpath, exist_ok=True)
        self.save_object(fullpath/Path(name + '.pkl'))
        self.save_as_table(fullpath/Path(name + '_table.xlsx'))
        self.save_as_vector(fullpath/Path(name + '_vector.xlsx'))
    
    @staticmethod
    def load(filepath):
        with open(filepath, 'rb') as f:
            field = pickle.load(f)
        return field
        
    def __add__(self, seq):
        a = self.vector
        b = seq.vector
        if len(a) < len(b):
            c = b.copy()
            c[:len(a)] += a
        else:
            c = a.copy()
            c[:len(b)] += b
        sum_seq = self.__copy__()
        sum_seq.vector = c
        sum_seq.vector = sum_seq.vector % self.p
        return sum_seq
        
        
class Experiment:
    def __init__(self, p, m_seq_lst, length_lst):
        if len(m_seq_lst) != len(length_lst):
            raise IndexError("Number of M-sequences should be equal to the number of lengths")
        
```

```python
def check_m_seq(m_seq):
    # число символов (N+1) / p
    for sym in range(1, m_seq.p):
        assert sum([1 for i in m_seq.vector if i == sym]) == int((m_seq.N+1) / m_seq.p)
    
    # число нулей (N+1) / p - 1
    assert sum([1 for i in m_seq.vector if i == 0]) == int((m_seq.N+1) / m_seq.p - 1)
    
check_m_seq(m_seq_1)    
check_m_seq(m_seq_2)    
check_m_seq(m_seq_3)
```

```python
def get_Li_metric(seq_1, seq_2):
    if seq_1.p != seq_2.p:
        raise IndexError("Sequences have different 'p' parameters: %s and %s" % (seq_1.p, seq_2.p))
        
    if seq_1.N != seq_2.N:
        raise IndexError("Sequences have different 'N' parameters: %s and %s" % (seq_1.N, seq_2.N))
        
    if len(seq_1.vector) != len(seq_1.vector):
        raise IndexError("Sequences should have equal lengths")
        
    res = 0
    p = seq_1.p
    for i in range(len(seq_1.vector)):
        diff = abs(seq_1.vector[i][0] - seq_2.vector[i][0])
        res += diff if diff <= p // 2 else p - diff
    return res

def get_Euclidian_metric(seq_1, seq_2):
    if seq_1.p != seq_2.p:
        raise IndexError("Sequences have different 'p' parameters: %s and %s" % (seq_1.p, seq_2.p))
        
    if seq_1.N != seq_2.N:
        raise IndexError("Sequences have different 'N' parameters: %s and %s" % (seq_1.N, seq_2.N))
        
    if len(seq_1.vector) != len(seq_1.vector):
        raise IndexError("Sequences should have equal lengths")
    
    _sum = 0
    j = complex(0, 1)
    p = seq_1.p
    pi = math.pi
    for i in range(len(seq_1.vector)):
        diff = cmath.exp(j * seq_1.vector[i][0] * 2 * pi / p) * \
               cmath.exp(j * (p-seq_2.vector[i][0]) * 2 * pi / p)
#         diff = abs(seq_1.vector[i][0] - seq_2.vector[i][0])
        _sum += diff
    res = 1 / N * _sum
    return math.sqrt(res.real ** 2 + res.imag ** 2)
    
# get_Li_metric(m_seq_1, m_seq_1.shift(15))
get_Euclidian_metric(m_seq_1, m_seq_1.shift(11))
```

```python
import ipywidgets as widgets
from ipywidgets import Layout, Button, Box, Text, Textarea, BoundedIntText, BoundedFloatText, Dropdown, IntText, Output, IntSlider

items_layout = Layout( width='auto')     # override the default width of the button to 'auto' to let the button grow

box_layout = Layout(display='flex',
                    flex_flow='column',
                    align_items='stretch',
                    border='none',
                    width='50%')


class Context(Text):
    def __init__(self):
        self.objects = dict()
    
    def append(self, name, value):
        change = {'name': 'value', 
         'old': 'Field_1', 
         'new': 'Field_2', 
         'owner': Text(value='Field_2', description='Field Name:'), 
         'type': 'change'}
        self.objects[name] = value
        self.notify_change(change)

class  FieldCreator:
    def __init__(self, context):
        
        self.context = context
#         self.upload_in = FileUpload(
#             accept='',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
#             multiple=False)
        
        self.title_in = widgets.HTML(value="<h2>Create GMW</b2>")
        
        self.field_name_in = Text(
            value='Field_1',
            description='Field Name:',
            disabled=False
        )
        
        self.p_in = Dropdown(
        options=[2, 3, 5, 7, 11],
        value=7,
        description='p:',
        disabled=False,
        )

        self.S_in = BoundedIntText(
            value=4,
            min=1,
            max=10,
            step=1,
            description='S:',
            disabled=False
        )

        self.m_in = BoundedIntText(
            value=2,
            min=1,
            max=10,
            step=1,
            description='m:',
            disabled=False
        )
        
        self.n_in = BoundedFloatText(
            value=self.S_in.value / self.m_in.value,
            description='n:',
            disabled=True
        )
        

        self.n_sum_in = IntText(
            value=100,
            min=1,
            step=1,
            description='n sum:',
            disabled=False
        )
        
        # shift (roll)
        self.A_in = IntText(
            value=0,
            step=1,
            description='A:',
            disabled=False
        )
        
        self.Ci_in = Text(
            value='2, 4, 6',
            description='Ci:',
            disabled=False
        )
        
        self.B_in = Text(
            value=''.join(['%sd%s ' % (val, idx) for idx, val in enumerate(self.Ci)]),
            description='B:',
            disabled=True
        )

        self.seed_in = Text(
            value='4, 0, 5, 5',
            description='NU:',
            disabled=False
        )
        
        self.create_in = Button(description="Create GMW")
        self.create_in.on_click(self.on_create)
        
        self.input_items = [v for k, v in self.__dict__.items() if k.endswith('_in')]
        for item in self.input_items: 
            item.observe(self.update, names='value')
        
    def update(self, change):
        self.n_in.value = self.S_in.value / self.m_in.value
        self.B_in.value = ''.join(['%sd%s ' % (val, idx) for idx, val in enumerate(self.Ci)])
    
    def on_create(self, b):
        f_name = self.field_name_in.value
        p = int(self.p_in.value)
        m = int(self.m_in.value)
        S = int(self.S_in.value)
        A = int(self.A_in.value)
        B = self.Ci  # yeah, need to rename it
        seed = self.seed
        m_seq = MSequnce(p, m, S, B, seed, A)
        self.context.append(f_name, m_seq)
    
    @property
    def Ci(self):
        return [int(num) for num in self.Ci_in.value.replace(' ','').split(',') if num.isdigit()]
    
    @property
    def seed(self):
        return [int(num) for num in self.seed_in.value.replace(' ','').split(',') if num.isdigit()]  
```

```python
class DisplayGMW:
    def __init__(self, context):
        self.context = context
        self.title_in = widgets.HTML(value="<h2>Show GMW</b2>")
        self.field_in = Dropdown(
            options=list(self.context.objects.keys()),
            description='GMW:',
            disabled=False,
        )
        self.output_in = Output()
        
        self.display_in = Button(description="Display")
        self.display_in.on_click(self.display)
        
        self.input_items = [v for k, v in self.__dict__.items() if k.endswith('_in')]

        self.context.observe(self.update, names='value')
        
    def update(self, change):
        self.field_in.options = list(self.context.objects.keys())
        
    def display(self, b):
        self.output_in.clear_output()
        with self.output_in:
            display(self.context.objects.get(self.field_in.value).show_table())
        
# display_GMW = DisplayGMW(context)
# box = Box(children=display_GMW.input_items, layout=box_layout)
# box
```

```python
class CompareGMW:
    def __init__(self, context):
        self.context = context
        
        self.title_in = widgets.HTML(value="<h2>Correllation</b2>")                
        self.max_Li = "Undefined"
        self.min_Li = "Undefined"
        self.max_Eu = "Undefined"
        self.min_Eu = "Undefined"

        self.li_y = []
        self.eu_y = []
        self.x = []

        self.field_1_in = Dropdown(
            options=list(self.context.objects.keys()),
            description='GMW 1:',
            disabled=False,
        )        
        self.field_2_in = Dropdown(
            options=list(self.context.objects.keys()),
            description='GMW 2:',
            disabled=False,
        )
        
        self.shift_in = IntText(
            value=-3,
            step=1,
            description='GMW 1 shift:',
            disabled=True
        )
        
        self.x_axis_in = IntText(
            value=10,
            step=1,
            description='x axis length:',
            disabled=True
        )
        
        self.stats_in = widgets.HTML(
            value=self.stats_report()
        )
        
        self.output_in = Output()
        
        self.display_in = Button(description="Plot metrics")
        self.display_in.on_click(self.display)
        
        self.filename_in = Text(
            value='metrics.xlsx',
            description='File name:',
            disabled=False
        )
        
        self.save_stats_in = Button(description="Save metrics")
        self.save_stats_in.on_click(self.save_metrics)
        
        self.input_items = [v for k, v in self.__dict__.items() if k.endswith('_in')]

        self.context.observe(self.update, names='value')
        self.shift_in.observe(self.display, names='value')
        self.x_axis_in.observe(self.display, names='value')
        self.filename_in.observe(self.display, names='value')
    
    def update(self, change):
        self.shift_in.disabled = False
        self.x_axis_in.disabled = False
        self.field_1_in.options = list(self.context.objects.keys())
        self.field_2_in.options = list(self.context.objects.keys())
        self.stats_in.value = self.stats_report()
    
    def stats_report(self):
        eu_str = "<h3>Euclidean metric:</h3>max: %s<br> min: %s<br>" % (self.max_Eu, self.min_Eu)
        li_str = "<h3>Li metric:</h3>max: %s<br> min: %s<br>" % (self.max_Li, self.min_Li)
        return li_str + eu_str
    
    def display(self, b):
        self.output_in.clear_output()
        with self.output_in:
            shift = self.shift_in.value
            x_length = self.x_axis_in.value
            m_seq_1 = self.context.objects[self.field_1_in.value]
            m_seq_2 = self.context.objects[self.field_2_in.value]
            y_Li = []
            y_Eu = []
            x = []
            for i in range(x_length):
                Di_Li = get_Li_metric(m_seq_1.shift(shift + i), m_seq_2.shift(0))
                Ri_Li = m_seq_1.N - 4 / m_seq_1.p * Di_Li
                Ri_Eu = get_Euclidian_metric(m_seq_1.shift(shift + i), m_seq_2.shift(0))
                x.append(i)
                y_Li.append(Ri_Li)
                y_Eu.append(Ri_Eu)
                
            self.max_Li = max(y_Li)
            self.min_Li = min(y_Li)
            self.max_Eu = max(y_Eu)
            self.min_Eu = min(y_Eu)

            self.li_y = y_Li
            self.eu_y = y_Eu
            self.x = x

            self.stats_in.value = self.stats_report()
                
            fig, (ax1, ax2) = plt.subplots(2)
            ax1.plot(x, y_Li)
            ax2.plot(x, y_Eu)
            plt.show()
    
    def save_metrics(self, b):
        metrics = {'Lambda': self.x,
         'Li': self.li_y,
         'Euclidean': self.eu_y}
        df = pd.DataFrame(metrics)
        df.to_excel(self.filename_in.value, index=False, header=True)
        
    
```

```python
class  FieldSum:
    def __init__(self, context):
        
        self.context = context
        
        self.title_in = widgets.HTML(value="<h2>PRS sum</b2>")
        
        self.field_name_in = Text(
            value='PRS_1',
            description='PRS name:',
            disabled=False
        )
        
        self.field_1_in = Dropdown(
            options=list(self.context.objects.keys()),
            description='GMW 1:',
            disabled=False,
        )        
        self.field_2_in = Dropdown(
            options=list(self.context.objects.keys()),
            description='GMW 2:',
            disabled=False,
        )
        
        self.create_in = Button(description="Create SUM")
        self.create_in.on_click(self.on_create)
        
        self.input_items = [v for k, v in self.__dict__.items() if k.endswith('_in')]
        
        self.context.observe(self.update, names='value')
        
    def update(self, change):
        self.field_1_in.options = list(self.context.objects.keys())
        self.field_2_in.options = list(self.context.objects.keys())
    
    def on_create(self, b):
        f_name = self.field_name_in.value
        seq_1 = self.context.objects[self.field_1_in.value]
        seq_2 = self.context.objects[self.field_2_in.value]
        seq = seq_1 + seq_2
        self.context.append(f_name, seq)
```

```python
class  FieldLoad:
    def __init__(self, context):
        
        self.context = context
        
        self.title_in = widgets.HTML(value="<h2>Load</b2>")
        
        self.folder_name_in = Text(
            value='experiments',
            description='Folder:',
            disabled=False
        )
        
        
        self.to_load_in = Dropdown(
            options=self.list_folder(),
            description='Name:',
            disabled=False if self.folder_exists() else True,
        )
        
        self.loaded_name_in = Text(
            value='Loaded field',
            description='Load as:',
            disabled=False if self.folder_exists() else True
        )
        
        self.load_btn_in = Button(
            description="Load",
            disabled=False if self.folder_exists() else True
        )
        self.load_btn_in.on_click(self.on_load)
        
        self.refresh_btn_in = Button(
            description="Refresh",
            disabled=False if self.folder_exists() else True
        )
        self.refresh_btn_in.on_click(self.update)
        
        self.input_items = [v for k, v in self.__dict__.items() if k.endswith('_in')]
        
        self.context.observe(self.update, names='value')
        self.folder_name_in.observe(self.update, names='value')
        
    def update(self, change):
        self.to_load_in.options = self.list_folder()
        self.to_load_in.disabled = False if self.folder_exists() else True
        self.loaded_name_in.disabled = False if self.folder_exists() else True
        self.load_btn_in.disabled = False if self.folder_exists() else True
#         self.field_2_in.options = list(self.context.objects.keys())
    
    def on_load(self, b):
        fielddir = Path(self.to_load_in.value)
        filepath = Path(self.folder_name_in.value)/fielddir/Path(fielddir.stem + '.pkl')
        seq = MSequnce.load(filepath)
        self.context.append(self.loaded_name_in.value, seq)
        
    def list_folder(self):
        try:
            return os.listdir(self.folder_name_in.value)
        except IOError:
            return ["No such folder"]
        
    def folder_exists(self):
        if os.path.exists(Path(self.folder_name_in.value)):
            return True
        else: 
            return False
```

```python
class  FieldSave:
    def __init__(self, context):
        
        self.context = context
        
        self.title_in = widgets.HTML(value="<h2>Save</b2>")
        
        self.folder_name_in = Text(
            value='experiments',
            description='Folder:',
            disabled=False
        )
        
        self.to_save_in = Dropdown(
            options=list(self.context.objects.keys()),
            description='Save:',
            disabled=False,
        )    
        
        self.saved_name_in = Text(
            value='',
            description='Save as:',
            disabled=False
        )
        
        self.save_btn_in = Button(
            description="Save",
            disabled=False
        )
        self.save_btn_in.on_click(self.on_save)
        
        self.save_all_btn_in = Button(
            description="Save All",
            disabled=False
        )
        self.save_all_btn_in.on_click(self.on_save_all)
        
        self.input_items = [v for k, v in self.__dict__.items() if k.endswith('_in')]
        
        self.context.observe(self.update, names='value')
        self.to_save_in.observe(self.update, names='value')
        
    def update(self, change):
        self.to_save_in.options = list(self.context.objects.keys())
        self.saved_name_in.value = self.to_save_in.value
    
    def on_save(self, b):
        name = self.saved_name_in.value
        filepath = self.folder_name_in.value
        self.context.objects[self.to_save_in.value].save_all(name, filepath)
    
    def on_save_all(self, b):
        for name in self.context.objects.keys():
            filepath = self.folder_name_in.value
            self.context.objects[name].save_all(name, filepath)
```
